讀取資料

```python
import numpy as np
import pandas as pd

file1 = 'db2020_hw5_ptt_mobilecomm.csv'
ptt20 = pd.read_csv(filepath_or_buffer=file1, error_bad_lines=False)
ptt20 = ptt20.dropna()
ptt20.index = range(0, 15231, 1)
ptt20
```

```python
file2 = 'db2020_hw5_dcard_3c.csv'
dcard20 = pd.read_csv(filepath_or_buffer=file2, error_bad_lines=False)
dcard20 = dcard20.dropna()
dcard20.index = range(0, 17745, 1)
dcard20
```

建立dataframe

```python
# 建立計分dataframe
models= {
    'name' : ['iPhone 12', 'iPhone 11', 'iPhone SE', 'Zenfone 7', 'Pixel 4', 'U20', 
              'Reno 4', 'C3', 'A21s', 'A71', 'A31', 'S20', 'Note 20', 'Xperia 5', 'X50'],
    'titlenum(ptt)' : [0] * 15,
    'contentnum(ptt)' : [0] * 15,
    'commentnum(ptt)' : [0] * 15,
    'discussion(ptt)' : [0] * 15
}

model_df = pd.DataFrame(models)

"""
各欄位依序為:
title數(ptt), content數(ptt), commentnum(ptt), 討論度(ptt)
"""

# 建立各機型名稱高相關詞彙
model_alias_dict = {
    'iPhone 12': ['i12', 'iphone12', 'iphone 12', 'iPhone12', 'iPHONE 12', 'iPhone 12'],
    'iPhone 11': ['i11', 'iphone11', 'iphone 11', 'iPhone11', 'iPHONE 11', 'iPhone 11'],
    'Zenfone 7': ['ZF7', 'zf7', 'Zenfone 7'],
    'Pixel 4': ['p4', 'P4', 'Pixel 4'],
    'Note 20': ['N20', 'Note20', 'Note 20'],
    'Xperia 5': ['Sony 5', 'Xperia 5'],
    'A21s': ['A21S', 'A21s', 'A21'],
    'S20': ['S20', 's20']
}

# 重新編排index
model_df.index = ['iPhone 12', 'iPhone 11', 'iPhone SE', 'Zenfone 7', 'Pixel 4', 'U20', 
              'Reno 4', 'C3', 'A21s', 'A71', 'A31', 'S20', 'Note 20', 'Xperia 5', 'X50']
model_df
```

分析標題、內容並統計結果

```python
for model in model_df['name']:
    # 各機型在ptt的四項指標
    title_ptt = 0
    content_ptt = 0
    commentnum_ptt = 0
    discussion_ptt = 0

    for i in range(15231):  # 過濾全部文章
        title = ptt20.at[i, 'title']
        content = ptt20.at[i, 'content']
        commentnum = ptt20.at[i, 'comment_count']
        
        if model in model_alias_dict:
            for alias in model_alias_dict[model]:  # 跑過所有別稱
                try:
                    if title.find(alias) != -1:
                        title_ptt += 1
                        commentnum_ptt += int(commentnum)

                    if content.find(alias) != -1:
                        content_ptt += 1
                        if title.find(alias) == -1:  # 只有內文有出現
                            commentnum_ptt += int(commentnum)

                # 避免未知錯誤
                except AttributeError:
                    continue

                except ValueError:
                    continue
        else:
            try:
                if title.find(model) != -1:
                    title_ptt += 1
                    commentnum_ptt += int(commentnum)
            
                if content.find(model) != -1:
                    content_ptt += 1
                    if title.find(model) == -1:  # 只有內文有出現
                        commentnum_ptt += int(commentnum)
            # 避免未知錯誤
            except AttributeError:
                continue

            except ValueError:
                continue

    discussion_ptt = 10*title_ptt + 5*content_ptt + commentnum_ptt  # 計算討論度
    
    # 將此機型資料貼到dataframe
    model_df.at[model, 'titlenum(ptt)'] = title_ptt
    model_df.at[model, 'contentnum(ptt)'] = content_ptt
    model_df.at[model, 'commentnum(ptt)'] = commentnum_ptt
    model_df.at[model, 'discussion(ptt)'] = int(discussion_ptt)

model_df
```

```python
model_df = model_df.drop('name', axis=1)
model_df = model_df.sort_values(['discussion(ptt)'], ascending=False)
model_df
```

以下為iphone12競品分析，建立需要的df和演算法

```python
# 競爭者計分dataframe
competitors= {
    'name' : ['iPhone 11', 'iPhone SE', 'Zenfone 7', 'Pixel 4', 'U20', 
              'Reno 4', 'C3', 'A21s', 'A71', 'A31', 'S20', 'Note 20', 'Xperia 5', 'X50'],
    '共同提及' : [0] * 14,
}

competitors_df = pd.DataFrame(competitors)

# 競爭者的高相關詞彙
competitors_alias_dict = {
    'iPhone 11': ['i11', 'iphone11', 'iphone 11', 'iPhone11', 'iPHONE 11', 'iPhone 11'],
    'Zenfone 7': ['ZF7', 'zf7', 'Zenfone 7'],
    'Pixel 4': ['p4', 'P4', 'Pixel 4'],
    'Note 20': ['N20', 'Note20', 'Note 20'],
    'Xperia 5': ['Sony 5', 'Xperia 5'],
    'A21s': ['A21S', 'A21s', 'A21'],
    'S20': ['S20', 's20']
}

# i12的高相關詞彙
i12_alias = ['i12', 'iphone12', 'iphone 12', 'iPhone12', 'iPHONE 12', 'iPhone 12']

# 重新編排index
competitors_df.index = ['iPhone 11', 'iPhone SE', 'Zenfone 7', 'Pixel 4', 'U20', 
              'Reno 4', 'C3', 'A21s', 'A71', 'A31', 'S20', 'Note 20', 'Xperia 5', 'X50']
# 清理資料
competitors_df = competitors_df.drop('name', axis=1)
competitors_df
```

```python
for competitor in competitors_df.index:
    for i in range(15231):
        article = ptt20.at[i, 'content']  # 每篇文章
        
        for i12 in i12_alias:
            if competitor in competitors_alias_dict:  # 競品有別稱
                for com in competitors_alias_dict[competitor]:
                    if article.find(i12) != -1 and article.find(com) != -1:  # 兩者被共同提及
                        competitors_df.at[competitor, '共同提及'] += 1
            else:  # 競品無別稱
                if article.find(i12) != -1 and article.find(competitor) != -1:  # 兩者被共同提及
                        competitors_df.at[competitor, '共同提及'] += 1

competitors_df
```

```python
competitors_df = competitors_df.sort_values(['共同提及'], ascending=False)
competitors_df
```
