```python
import numpy as np
import pandas as pd

file1 = 'db2020_hw5_ptt_mobilecomm.csv'
ptt20 = pd.read_csv(filepath_or_buffer=file1, error_bad_lines=False)  # 用pandas開啟csv檔
ptt20 = ptt20.dropna()  # 將空值去除
ptt20.index = range(0, 15231, 1)  # 重新編號
ptt20
```

```python
tncs = ptt20[['title', 'content']]  # 挑出需要的資料
ask = []  # 用來存放問機文的list

tncs

for i in range(15231):  # 過濾所有文章並從中挑出問機文
    if tncs.at[i, 'title'].find('問機') != -1:
        ask.append([tncs.at[i, 'title'], tncs.at[i, 'content']])

# 建構問機文的dataframe
ask_df = pd.DataFrame(ask)
ask_df.columns = ['title', 'content']
ask_df
```

```python
# 各種方面的計分字典
aspects = {'拍照': 0, '音效': 0, '續航': 0, '外觀': 0, 
           'CP值': 0, '效能': 0}

# 同義詞字典
aspects_syn = {
    '拍照': ['拍照', '鏡頭', '夜拍', '照相'],
    '音效': ['音效', '聽歌', '聲音', '通話'],
    '續航': ['續航', '充電', '電池', '快充'], 
    '外觀': ['外觀', '螢幕', '尺寸', '顏色'],
    'CP值': ['CP值', '優惠', '便宜', '特價'],
    '效能': ['效能', '流暢', '速度', '性能', '不卡']
}

# 須出現2次才能算1 mention(因為發文格式)
two_times = ['拍照', '續航', '尺寸', '螢幕', '效能']  
```

```python
for aspect in aspects:  # 跑每個方面
    mention = 0  # 提及次數(分數)

    for i in range(1513):  # 跑每篇文章
        article = ask_df.at[i, 'content']
        
        # 跑同義詞
        for syn in aspects_syn[aspect]:
            if syn not in two_times:
                if article.count(syn) >= 1:
                    mention += 1
            else:
                if article.count(syn) >= 2:
                    mention += 1
    print(mention)
    aspects[aspect] = mention
```

```python
aspect
aspects_ser = pd.Series(aspects)
aspects_ser = aspects_ser.sort_values(ascending=False)
aspects_ser
```

以下為iphone12和三星s20的共同提及優缺點關鍵字比較

```python
# 建立只有title和content的df
tnc_df = ptt20[['title', 'content', 'sentiment']]

# i12計分字典
goods_iPhone12 = {'拍照': 0.0, '音效': 0.0, '續航': 0.0, '外觀': 0.0, 'CP值': 0.0, '效能': 0.0}
bads_iPhone12 = {'拍照': 0.0, '音效': 0.0, '續航': 0.0, '外觀': 0.0, 'CP值': 0.0, '效能': 0.0}

# 優缺點高相關詞彙字典
aspects_syn = {
    '拍照': ['拍照', '鏡頭', '夜拍', '照相'],
    '音效': ['音效', '聽歌', '聲音', '通話'],
    '續航': ['續航', '充電', '電池', '快充'], 
    '外觀': ['外觀', '螢幕', '尺寸', '顏色'],
    'CP值': ['CP值', '優惠', '便宜', '特價'],
    '效能': ['效能', '流暢', '速度', '性能', '不卡']
}

# i12高相關詞彙
i12_alias = ['i12', 'iphone12', 'iphone 12', 'iPhone12', 'iPHONE 12', 'iPhone 12']
tnc_df
```

```python
for i12 in i12_alias:
    for i in range(15231):
        title = tnc_df.at[i, 'title']
        content = tnc_df.at[i, 'content']
        sentiment = tnc_df.at[i, 'sentiment']
        
        if title.find(i12) == -1 and content.find(i12) == -1:  # 標題文章都沒提到
            continue

        else:
            for aspect in goods_iPhone12:
                for syn in aspects_syn[aspect]:
                    if content.count(syn) >= 1 or title.count(syn) >= 1:  # 文章有提及
                        if sentiment == 'P':
                            goods_iPhone12[aspect] += 1
                        elif sentiment == 'N':
                            bads_iPhone12[aspect] += 1
                        
                        else:
                            goods_iPhone12[aspect] += 0.5
                            bads_iPhone12[aspect] += 0.5

print('i12正面分數', goods_iPhone12, sep=':')
print('i12負面分數', bads_iPhone12, sep=':')
```
